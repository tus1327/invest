```kotlin
%use dataframe
%use kandy
```

```kotlin
val years = 20
val start = 3000

val year = (0..years).toList()

// 연봉인상률
val t_rate = year.map {
    when {
        it <= 5 -> 0.1
        it <= 10 -> 0.08
        it <= 15 -> 0.06
        it <= 20 -> 0.04
        else -> 0.03
    }
}

// 수익률
val i_rate = 0.07

val principal = year.zip(t_rate).runningFold(start.toDouble()) { acc, (s, r) ->
    acc * (1 + r)
}.dropLast(1)

val db = principal.map { (it / 12.0) }.mapIndexed { i, d -> d * (i + 1) }

val dc = principal.map { (it / 12.0) }.runningFold(0.0) { acc, d -> (acc * (1+ i_rate)) + d }.drop(1)

val dbc = db.take(14) + principal.map { it / 12.0 }.drop(14).runningFold(db[13]) { acc, d -> (acc * (1 + i_rate)) + d }
    .drop(1)

val df = dataFrameOf(
    "year" to year.map { it + 1 },
    "연봉" to principal.map { it.toInt() },
    "DB" to db.map { it.toInt() },
    "DC" to dc.map { it.toInt() },
    "DB>DC" to dbc.map { it.toInt() },
)

df
```

```kotlin
df.plot {
    path {
        x("year")
        y("연봉")
    }
    path {
        x("year")
        y("DB")
        color = Color.RED
    }
    path {
        x("year")
        y("DC")
        color = Color.ORANGE
    }
    path {
        x("year")
        y("DB>DC")
        color = Color.YELLOW

    }
}
```
