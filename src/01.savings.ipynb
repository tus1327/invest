### Savings

```kotlin
%use kandy
%use dataframe
```

```kotlin
fun savings(
    months: Int,
    rate: Double,
    list: List<Int>
): DataFrame<*> {
    val 저축액 = list.runningReduce { acc, i -> acc + i }
    val 이자 = 저축액.map { it * rate / 12 }
    val 이자누적 = 이자.runningReduce { acc, i -> acc + i }
    return dataFrameOf(
        "월" to (1..months).toList(),
        "월납입" to list,
        "저축" to 저축액,
        "이자" to 이자,
        "이자누적" to 이자누적,
        "총가치" to 저축액.zip(이자누적).map { (s, i) -> s + i }.map { it.toInt() },
    )
}
```

```kotlin
savings(12, 0.03, List(12) { 100000 })
```

```kotlin
val df = savings(12, 0.02, List(12) { 10 })
df.plot {
    area {
        x("월")
        y("총가치")
        fillColor = Color.BLUE

    }
    area {
        x("월")
        y("저축")
        fillColor = Color.RED
    }

//    path {
//        x("월")
//        y("저축")
//    }
    layout { size = 1200 to 600 }
}
```
