```kotlin
fun calcTax(퇴직금: Int, 근속연수: Int): Int {

    val 근속연수공제 = if (근속연수 <= 5) {
        근속연수 * 100
    } else if (근속연수 <= 10) {
        500 + (근속연수 - 5) * 200
    } else if (근속연수 <= 20) {
        1500 + (근속연수 - 10) * 250
    } else {
        4000 + (근속연수 - 20) * 300
    }

//println("근속연수공제 : $근속연수공제")

    val 환산급여 = (퇴직금 - 근속연수공제) * 12 / 근속연수
//println("환산급여 : $환산급여")

    val 환산급여공제 = if (환산급여 <= 800) {
        환산급여
    } else if (환산급여 <= 7000) {
        (800 + (환산급여 - 800) * 0.6).toInt()
    } else if (환산급여 <= 1_0000) {
        (4520 + (환산급여 - 7000) * 0.55).toInt()
    } else if (환산급여 <= 3_0000) {
        (6170 + (환산급여 - 1_0000) * 0.45).toInt()
    } else {
        (15170 + (환산급여 - 3_0000) * 0.35).toInt()
    }

//println("환산급여공제 : $환산급여공제")
    val 과세표준 = 환산급여 - 환산급여공제

//println("과세표준 : $과세표준")

    val 세율 = if (과세표준 <= 1400) {
        0.06
    } else if (과세표준 <= 5000) {
        0.15
    } else if (과세표준 <= 8800) {
        0.24
    } else if (과세표준 <= 15000) {
        0.35
    } else if (과세표준 <= 30000) {
        0.38
    } else if (과세표준 <= 50000) {
        0.40
    } else if (과세표준 <= 100000) {
        0.42
    } else {
        0.45
    }

    val 누진공제액 = if (과세표준 <= 1400) {
        0
    } else if (과세표준 <= 5000) {
        126
    } else if (과세표준 <= 8800) {
        576
    } else if (과세표준 <= 15000) {
        1544
    } else if (과세표준 <= 30000) {
        1994
    } else if (과세표준 <= 50000) {
        2594
    } else if (과세표준 <= 100000) {
        3594
    } else {
        6594
    }

    val 환산산출세액 = (과세표준 * 세율) - 누진공제액
//println("환산산출세액 : ${환산산출세액.toInt()}")

    val 산출세액 = 환산산출세액 / 12 * 근속연수
    val 지방소득세 = 산출세액 * 0.1
    val 세금 = 산출세액 + 지방소득세
    val 세후퇴직금 = 퇴직금 - 세금

//    println("산출세액 : ${산출세액.toInt()}")
//    println("세후퇴직금 : ${세후퇴직금.toInt()}")

    return 세금.toInt()
}
```

```kotlin
fun calcTaxPr(퇴직금: Int, 근속연수: Int) {
    println("퇴직금 : $퇴직금")
    println("근속연수 : $근속연수")
    val 세금 = calcTax(퇴직금, 근속연수)
    val 세후퇴직금 = 퇴직금 - 세금
    val 최종세율 = 세금.toDouble() / 퇴직금.toDouble()

    println("세금 : ${세금}")
    println("세후퇴직금 : ${세후퇴직금.toInt()}")
    println("최종세율 : ${String.format("%2.2f", 최종세율 * 100f)}%")
}
```

```kotlin
val list = listOf(
    Pair(1_0000, 1),
    Pair(1_0000, 5),
    Pair(1_0000, 10),
    Pair(1_0000, 20),
    Pair(3_0000, 1),
    Pair(3_0000, 5),
    Pair(3_0000, 10),
    Pair(3_0000, 20),
)

val 퇴직금 = list.map { it.first }
val 근속연수 = list.map { it.second }
val 세금 = list.map { calcTax(it.first, it.second) }
val 세후퇴직금 = 세금.zip(퇴직금) { tax, total -> total - tax }
val 최종세율 = 세금.zip(퇴직금) { tax, total -> tax.toDouble() / total.toDouble() }.map {
    "${String.format("%2.2f", it * 100f)}"
}

dataFrameOf(
    "퇴직금" to 퇴직금,
    "근속연수" to 근속연수,
    "세후퇴직금" to 세후퇴직금,
    "세금" to 세금,
    "최종세율" to 최종세율,
)
```
