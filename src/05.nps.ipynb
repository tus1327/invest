```kotlin
val years = (60..100).toList()
val 월연금수령액 = 100

fun 조정계수(t: Int) : Double {
    return if (t < 0) {
        t * 0.06
    } else if (t == 0) {
        0.0
    } else {
        t * 0.072
    }
}

fun 연금수령월(t: Int): List<Double> {
    return years.map {
        if (it >= t + 65) {
            월연금수령액 * (1 + 조정계수(t))
        } else {
            0.0
        }
    }
}

fun 연금수령(t: Int, irate: Double = 0.0): List<Double> {
    return 연금수령월(t).runningFold(0.0) { acc, value -> (acc * (1+irate)) + value * 12 }.drop(1)
}


val list = listOf(-5, 0, 5)

list.forEach {
    val v = 월연금수령액 * (1 + 조정계수(it))
    println("t=${it + 65}, $v")
}

val 할인율 = 0.03

val m = mapOf("y" to years) + list.map {
    "$it" to 연금수령(it, 할인율)
}

val df = m.toDataFrame()
df.plot {
    list.onEach { l ->
        path {
            x("y")
            y("$l")
            color = when {
                l == -5 -> Color.RED
                l == 0 -> Color.BLUE
                l == 5 -> Color.PURPLE
                else -> Color.WHITE
            }
        }
    }
    layout { size = 1200 to 800 }
}
```
